# Zomato Bangalore Restaurants

## Import the essential libraries

```python

import pandas as pd #pandas for Data cleaning
import numpy as np #numpy for Numerical calculations
import matplotlib.pyplot as plt #matplot and seaborn for Data Visualization
import seaborn as sns

# Set some display options for better viewing
pd.set_option('display.max_columns', None)
plt.style.use('seaborn-v0_8-whitegrid')

print("Libraries imported successfully!")
```

##  Data Collection

```python
df = pd.read_csv('zomato.csv') # Load the dataset 

print("First 5 rows of the dataset:")
display(df.head())

print("Last 5 rows of the dataset: ")
display(df.tail())

print("\nDataset Information:")
df.info()


print("\nMissing values in each column:") # Check for the number of missing values in each column
print(df.isnull().sum())
```

##  Data Cleaning / Preprocessing

```python
# --- Complete and Correct Cleaning for the 'rate' Column ---

# 1. Define the function to handle all known string formats
def clean_rate(rate):
    if isinstance(rate, str):
        rate = rate.strip() # Remove leading/trailing spaces
        if rate == 'NEW' or rate == '-':
            return np.nan # Return a missing value
        else:
            # Handle formats like "4.1/5"
            return float(rate.split('/')[0])
    return rate

# 2. Apply this custom function first
print("Applying custom cleaning to 'rate' column...")
df['rate'] = df['rate'].apply(clean_rate)

# 3. As a final safeguard, force the column to a numeric type
# This will catch any other weird values the function missed
df['rate'] = pd.to_numeric(df['rate'], errors='coerce')

# 4. Now, calculate the median and fill any remaining missing values
median_rate = df['rate'].median()
df['rate'] = df['rate'].fillna(median_rate)

print("'rate' column has been successfully cleaned and is ready for analysis.")
print(df['rate'].describe())
```

```python
# Clean and rename the 'approx_cost(for two people)' column
df.rename(columns={'approx_cost(for two people)': 'cost_for_two'}, inplace=True)

# FIXED: First, ensure the column is a string, then remove the comma, then convert to float.
df['cost_for_two'] = df['cost_for_two'].astype(str).str.replace(',', '').astype(float)

# Calculate the median cost
median_cost = df['cost_for_two'].median()

# Fill any missing values with the median
df['cost_for_two'] = df['cost_for_two'].fillna(median_cost)

print("'cost_for_two' column cleaned and filled.")
```

```python
df = df.dropna(subset=['rest_type', 'cuisines']) #Droping the row with missing rest_type and cuisiness.Droping 45 and 227 entries from more than 50K will not affect the Datainsight
print("Rows with missing 'rest_type' or 'cuisines' have been dropped.")

print("\nMissing values after cleaning:")
print(df.isnull().sum())
```


## Data Visualization


### Top 10 type of restaurants are most likely to find in Bengaluru

```python
plt.figure(figsize=(12, 6)) #standard size of figure for visualization
sns.countplot(y='rest_type', data=df, order=df['rest_type'].value_counts().index[:10], hue='rest_type', legend=False)
plt.title('Top 10 Most Common Restaurant Types in Bengaluru', fontsize=16)
plt.xlabel('Number of Restaurants', fontsize=12)
plt.ylabel('Restaurant Type', fontsize=12) #
plt.show()
```

### Top 10 Most Popular Cuisines

```python

plt.figure(figsize=(12, 6))
cuisine_counts = df['cuisines'].value_counts()
# FIXED: Updated sns.barplot to use 'hue' and 'legend=False' for consistency.
sns.barplot(y=cuisine_counts.index[:10], x=cuisine_counts.values[:10], hue=cuisine_counts.index[:10], legend=False)
plt.title('Top 10 Most Popular Cuisines in Bengaluru', fontsize=16)
plt.xlabel('Number of Restaurants', fontsize=12)
plt.ylabel('Cuisine', fontsize=12)
plt.show()
```

```python
# Relationship between Restaurant Rating and Cost for Two
plt.figure(figsize=(10, 6))
sns.scatterplot(x='rate', y='cost_for_two', data=df, alpha=0.5)
plt.title('Relationship between Rating and Cost for Two', fontsize=16)
plt.xlabel('Rating (out of 5)', fontsize=12)
plt.ylabel('Approx. Cost for Two (INR)', fontsize=12)
plt.show()
```

###  Distribution of Restaurant Ratings

```python

plt.figure(figsize=(10, 6))
sns.histplot(df['rate'], kde=True, bins=20, color='coral')
plt.title('Distribution of Restaurant Ratings', fontsize=16)
plt.xlabel('Rating (out of 5)', fontsize=12)
plt.ylabel('Frequency', fontsize=12)

#Average PeakRating
peak_rating = df['rate'].mode().iloc[0]

# Add a vertical line at the peak's location
plt.axvline(peak_rating, color='purple', linestyle='--', 
            label=f'Peak (Mode): {peak_rating:.2f}')

plt.legend()

plt.show()
```

### Location-Based Analysis in Bengaluru city

```python

print("\nAnalyzing by Location...")
# Find the top 10 locations with the highest average ratings
top_locations = df.groupby('location')['rate'].mean().sort_values(ascending=False).head(10)

print("\nTop 10 Highest-Rated Locations:")
print(top_locations)

# Visualize the top locations
plt.figure(figsize=(12, 6))
sns.barplot(x=top_locations.values, y=top_locations.index, hue=top_locations.index, legend=False, palette='cool')
plt.title('Top 10 Highest-Rated Locations in Bengaluru', fontsize=16)
plt.xlabel('Average Rating', fontsize=12)
plt.ylabel('Location', fontsize=12)
plt.show()
```

### Impact of rating by type of odering

```python
print("\nAnalyzing the Impact of Online Ordering...")
online_order_impact = df.groupby('online_order')['rate'].mean()

print("\nAverage rating by Online Order availability:")
print(online_order_impact)

# Visualize the impact of online ordering
plt.figure(figsize=(6, 5))
sns.barplot(x=online_order_impact.index, y=online_order_impact.values, hue=online_order_impact.index, legend=False, palette='autumn')
plt.title('Impact of Online Ordering on Ratings', fontsize=16)
plt.xlabel('Offers Online Ordering', fontsize=12)
plt.ylabel('Average Rating', fontsize=12)
plt.show()
```

#### Cuisine and Cost Analysis

```python

print("\nAnalyzing Cost by Cuisine...")
# This analysis can be slow because of the large number of unique cuisine combinations
# We will filter out cuisines with less than 50 restaurants for a more stable result
cuisine_counts = df['cuisines'].value_counts()
common_cuisines = cuisine_counts[cuisine_counts >= 50].index
df_common_cuisines = df[df['cuisines'].isin(common_cuisines)]

most_expensive_cuisines = df_common_cuisines.groupby('cuisines')['cost_for_two'].mean().sort_values(ascending=False).head(10)

print("\nTop 10 Most Expensive Cuisines (among common cuisines):")
print(most_expensive_cuisines)

# Visualize the most expensive cuisines
plt.figure(figsize=(10, 5))
sns.barplot(x=most_expensive_cuisines.values, y=most_expensive_cuisines.index, hue=most_expensive_cuisines.index, legend=False, palette='magma')
plt.title('Top 10 Most Expensive Cuisines', fontsize=16)
plt.xlabel('Average Cost for Two (INR)', fontsize=12)
plt.ylabel('Cuisine', fontsize=12)
plt.show()
```

### Pie Chart for Online Order Availability

```python
# Get the counts for each category
online_order_counts = df['online_order'].value_counts()

# Create the pie chart
plt.figure(figsize=(5, 5))
plt.pie(online_order_counts, 
        labels=online_order_counts.index, 
        autopct='%1.1f%%', 
        startangle=90,
        colors=['#66b3ff','#99ff99'])

plt.title('Proportion of Restaurants Offering Online Ordering', fontsize=16)
plt.axis('equal') 
plt.show()
```

### Analysis of Average Cost by Location

```python

avg_cost_by_location = df.groupby('location')['cost_for_two'].mean().sort_values(ascending=False)

most_expensive = avg_cost_by_location.head(10)

least_expensive = avg_cost_by_location.tail(10)

print("--- Top 10 Most Expensive Locations for Dining ---")

most_expensive_rounded = most_expensive.round(2)
print(most_expensive_rounded)


plt.figure(figsize=(12, 7))
sns.barplot(x=most_expensive.values, y=most_expensive.index, hue=most_expensive.index, legend=False, palette='YlOrRd_r')
plt.title('Top 10 Most Expensive Locations for Dining', fontsize=16)
plt.xlabel('Average Cost for Two (INR)', fontsize=12)
plt.ylabel('Location', fontsize=12)
plt.show()


print("\n--- Top 10 Least Expensive Locations for Dining ---")
least_expensive_rounded = least_expensive.round(2)
print(least_expensive_rounded)



plt.figure(figsize=(12, 7))
# Sort ascending for this plot for better visual comparison
least_expensive_sorted = least_expensive.sort_values(ascending=True)
sns.barplot(x=least_expensive_sorted.values, y=least_expensive_sorted.index, hue=least_expensive_sorted.index, legend=False, palette='GnBu_r')
plt.title('Top 10 Least Expensive Locations for Dining', fontsize=16)
plt.xlabel('Average Cost for Two (INR)', fontsize=12)
plt.ylabel('Location', fontsize=12)
plt.show()
```
